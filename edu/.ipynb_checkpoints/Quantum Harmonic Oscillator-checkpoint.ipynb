Enery Eigenasis

Any observable operators in their own eigenbasis are diagonal, the diagonal entries being the eigenvalues. $<H>_{mn}=<\psi_m|H|\psi_n>=<\psi_m|E_n|\psi_n>=E_n \delta_{m,n}$.

The second quantization expression in this form is in the energy eigenbasis. This is readily seen from: $<\psi_m|H|\psi_n>=<\psi_m|\hbar \omega (a^\dagger a + \frac{1}{2})|\psi_n>=\delta_{m,n} \hbar \omega (n+\frac{1}{2})= \delta_{m,n} E_n$

# Quantum Harmonic Oscillator

### Hamiltonian

We consider the one dimensional SHO, which has a potential of the form $V(x) = \frac{k}{2} x^2$.

$$H= \frac{\hat{p}^2}{2m} + \frac{k}{2} \hat{x}^2$$

We are working in the position basis. The usual substitutions follow: $X->x$, $P->-i\hbar \frac{d}{dx}$, $|E>->\psi_E(x)$.

$$H= -\frac{\hbar^2}{2m}\frac{\partial^2}{\partial^2 x} + \frac{k}{2} \hat{x}^2$$

And it is common to express the spring constant as $k=m \omega^2$.

$$H= -\frac{\hbar^2}{2m}\frac{\partial^2}{\partial^2 x} + \frac{m \omega^2}{2} \hat{x}^2$$

### Commutations

$$[x,p]=i \hbar$$
$$[x,p^2]=2i \hbar p$$
$$[x^2,p^2]=4i\hbar p x - 2\hbar^2 I$$

Note that the canonical commutation relation $[x,p]=i \hbar$ is what brings about the zero point fluctuations in the system (and the nonzero energy ground state).

##### -Hamiltonian

$$[x,H]=[x,\frac{p^2}{2m}]=i \hbar \frac{p}{m}$$

##### -Translation

Translation operator $\hat{T}(l)=e^{-i \frac{pl}{\hbar}}$. Generates shifts in position.

$$[x,T(l)]=l T(l)$$

Boost operator $\hat{K}(o)=e^{i \frac{xo}{\hbar}}$. Generates shifts in momentum.

$$[x,T(l)]=l T(l)$$

##### -3D Case

$$[x_i,p_j]=i \hbar \delta_{i,j}$$
$$[x_i,x_j]=0$$
$$[p_i,p_j]=0$$


### Eigenfunctions

$$\psi_n(x) = \frac{1}{(2^n n! \sqrt{\pi})^{1/2}} H_n(x) e^{-x^2 /2}$$

Where the H function is the nth Hermite Polynomial. The Hermite polynomials have a useful recursion relation that can be found in Shankar-7.3.24.

### Eigenvalues

$$E_n = \hbar \omega (n+\frac{1}{2})$$

We can consider the energy eigenvalues in terms of n quanta. We pretend there exist fictitious particles, each with energy $\hbar \omega$, such that an increase in energy corresponds to increasing the number of quanta.

### Expectation Values

$$<n|x|n>=<x>=<p>=0$$
$$<x^2>=\frac{3 \hbar}{2m \omega}$$
$$<p^2>=\frac{3m \omega \hbar}{2}$$

# 2nd Quantization

### Hamiltonian

$$H= \hbar \omega(\hat{a}^\dagger \hat{a} +\frac{1}{2})$$

The second quantization expression in this form is in the energy eigenbasis, as opposed to the x basis above. It is diagonalized in terms of the Number operator ($\hat{N}=\hat{a}^\dagger \hat{a}$) where N has the eigenvalues {1,2,...}, and each eigenvalue corresponds to a stae $|n>$.

### Ladder Operators

Introducing the raising $a^\dagger$ and lowering $a$ operators.

$$a= \sqrt{\frac{m \omega}{2 \hbar}}(\hat{x} + \frac{i}{m \omega} \hat{p})$$
$$a^\dagger= \sqrt{\frac{m \omega}{2 \hbar}}(\hat{x} - \frac{i}{m \omega} \hat{p})$$

$$a^\dagger |n> = \sqrt{n+1}|n+1>$$
$$a |n> = \sqrt{n}|n-1>$$

Number operator N (counts quanta):

$$N=a^\dagger a$$
$$N|n>=n|n>$$

### 2nd Quantization Commutations

$$[a,a^\dagger]=I$$
$$[a,(a^\dagger)^n]=n(a^\dagger)^{n-1}$$
$$[N,a^\dagger]=a^\dagger$$
$$[N,a]=-a$$
$$[H,a]=-\hbar \omega a$$
$$[H,a^\dagger]=\hbar \omega a^\dagger$$

### Relations

We can express all normalized eigenvectors $|n>$ in terms of the ground state $|0>$.

$$|n>= \frac{(a^\dagger)^n}{(n!)^1/2} |0>$$

# Relevant Part

$$H= \hbar \omega(N +\frac{1}{2})$$

In the energy basis, N and H are diagonal while $a, a^\dagger, q, p$ connect each energy eigenstate to the states immediatly above or below it.

$$N -> 
\begin{pmatrix}
0 & 0 & 0 & ...\\
0 & 1 & 0 & ...\\
0 & 0 & 2 & ...
\end{pmatrix}$$

$$H -> 
\hbar \omega
\begin{pmatrix}
1/2 & 0 & 0 & ...\\
0 & 3/2 & 0 & ...\\
0 & 0 & 5/2 & ...
\end{pmatrix}$$

$$a -> 
\begin{pmatrix}
0 & \sqrt{1} & 0 & 0 & ...\\
0 & 0 & \sqrt{2} & 0 & ...\\
0 & 0 & 0 & \sqrt{3} & ...
\end{pmatrix}$$

$$a^\dagger -> 
\begin{pmatrix}
0 & 0 & 0 & ...\\
\sqrt{1} & 0 & 0 & ...\\
0 & \sqrt{2} & 0 & ...\\
0 & 0 & \sqrt{3} & ...
\end{pmatrix}$$

$$x -> 
\begin{pmatrix}
0 & \sqrt{1/2} & 0 & ...\\
\sqrt{1/2} & 0 & \sqrt{2/2} & ...\\
0 & \sqrt{2/2} & 0 & ...\\
0 & 0 & \sqrt{3/2} & ...
\end{pmatrix}$$

$$p -> 
\begin{pmatrix}
0 & -i\sqrt{1/2} & 0 & ...\\
i\sqrt{1/2} & 0 & -i\sqrt{2/2} & ...\\
0 & i\sqrt{2/2} & 0 & ...\\
0 & 0 & i\sqrt{3/2} & ...
\end{pmatrix}$$
